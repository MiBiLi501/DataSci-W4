```python
# global column, movieDataSet
with open("movies_metadata.csv") as fileObj:
    reader_obj = csv.reader(fileObj)

    
    column = [attribute for attribute in reader_obj.__next__()]
    rawMovieDataSet = {attribute : [] for attribute in column}

    for row in reader_obj:
        for data in zip(column, row):
            rawMovieDataSet[data[0]].append(data[1])
pd.DataFrame(rawMovieDataSet)
```

```python
re.findall("Disney", str(rawMovieDataSet))
```

```python
id = []
name = []
poster_path = []
backdrop_path = []


for collection in rawMovieDataSet["belongs_to_collection"]:
    if len(collection) == 0:
        # id.append(None)
        # name.append(None)
        # poster_path.append(None)
        # backdrop_path.append(None)
        continue
    collectionProccessed = eval(collection)
    
    id.append(collectionProccessed['id'])
    name.append(collectionProccessed['name'])
    poster_path.append(collectionProccessed['poster_path'])
    backdrop_path.append(collectionProccessed['backdrop_path'])

collectionDataset = {"id" : id, "name" : name, "poster path" : poster_path, "backdrop path" : backdrop_path}

collectionDf = pd.DataFrame(collectionDataset).astype({"id" : int})
collectionDf.info()
collectionDf
```
